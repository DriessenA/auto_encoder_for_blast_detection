```python
import pandas as pd
import numpy as np
import dotenv
import mlflow
import seaborn as sns
import matplotlib.pyplot as plt
from utils_preproc.utils import best_score_to_param_setting_number
```

```python
dotenv.load_dotenv("/u/adr/.secrets/mlflow_server.env", override=True)
client = mlflow.tracking.MlflowClient(tracking_uri="http://mlflow.adr-services.zc2.ibm.com:5000")
```

```python
exp_id = client.get_experiment_by_name("sklearn_L1O_Error2").experiment_id
runs = client.search_runs(experiment_ids=exp_id, max_results=3500)
error_runs_df = pd.DataFrame([{**run.data.params, **run.data.metrics, **dict(run.info)} for run in runs])
error_runs_df["param_setting"] = error_runs_df["param_setting"].astype(int)
error_runs_df = error_runs_df[error_runs_df["param_setting"]<200]
error_runs_df.shape
```

```python
error_runs_df.groupby("param_setting").size()[error_runs_df.groupby("param_setting").size()<16]
```

```python
exp_id = client.get_experiment_by_name("sklearn_L1O_Expression2").experiment_id
runs = client.search_runs(experiment_ids=exp_id, max_results=3500)
expr_runs_df = pd.DataFrame([{**run.data.params, **run.data.metrics, **dict(run.info)} for run in runs])
expr_runs_df.shape
```

```python
expr_runs_df.groupby("param_setting").size()[expr_runs_df.groupby("param_setting").size()<16]
```

```python
CV_loops = set([str(i) for i in range(16)])
CV_loops.difference(set(list(expr_runs_df[expr_runs_df["param_setting"]=="87"]["CV_loop"])))
```

```python
remove_cols = ["experiment_name", "mlflow_uri", "mlflow_artifacts_uri", "env_fpath", "end_time", 
               "experiment_id", "lifecycle_stage", "run_uuid", "start_time", "status", "user_id",
              "artifact_uri", "sklearn_data.dataset", "sklearn_data.dataset_params.val_file", 
               "sklearn_data.dataset_params.test_file",
               "sklearn_data.dataset_params.train_file", "sklearn_data.dataset_params.random_state",
              "sklearn_data.dataset_params.num_workers", "sklearn_data.dataset_params.input_cols",
               "sklearn_data.dataset_params.batch_size",
              "sklearn_data.dataset_params.target", "checkpoint.monitor",
               "checkpoint.every_n_epochs", "checkpoint.save_top_k"]
group_cols = ["models.model"]+[col for col in error_runs_df.columns if col.startswith("models.model_params") ]
```

```python
error_runs_df  = error_runs_df.loc[:,~(error_runs_df.columns.isin(remove_cols))]
error_avg_runs = error_runs_df.groupby(group_cols, dropna=False).mean().reset_index(drop=False)

expr_runs_df  = expr_runs_df.loc[:,~(expr_runs_df.columns.isin(remove_cols))]
expr_avg_runs = expr_runs_df.groupby(group_cols, dropna=False).mean().reset_index(drop=False)

avg_runs = pd.concat([error_avg_runs, expr_avg_runs], keys=["error", "expr"], names=["input", "row_id"])
avg_runs.reset_index("row_id", drop=True, inplace=True)
avg_runs.reset_index("input", drop=False, inplace=True)
avg_runs.sort_values("val_accuracy_score", ascending=False).reset_index(drop=True).iloc[:35, :]
```

```python
sns.barplot(data=avg_runs, x="models.model", y="val_accuracy_score", hue="input")
```
